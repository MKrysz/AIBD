
## Zadanie 3
Zdefiniuj poniższą funkcję  i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$ 

1. $x>-1 oraz x<1$
2. $x>-6 oraz x<6$
3. $x>0 oraz x<5$

Wspierając się dokumentacją [Matplotlib](https://matplotlib.org/). Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

```python
def f(x):
    return x**2 + 5

x_list =[np.linspace(-1,1),np.linspace(-6,6),np.linspace(0,5)]
titles = ["od -1 do 1", "od -6 do 6", "od 0 do 5"]

for i, x in enumerate(x_list):
    y = f(x)
    plt.plot(x, y)
    plt.xlabel("arguments")
    plt.ylabel("values")
    plt.title("Przykładowa parabola z argumentami " + titles[i])
    plt.legend("f")
    plt.show()
```
